# Create Bar Charts

```python
# Ensure plots are displayed inline within the Jupyter Nookbook directly below
# the code cell that produced it
%matplotlib inline
```

```python
# Import dependencies
# From matplotlib import pyplot and give it the alias plt
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## MATLAB Method

### Vertical Bar 

```python
# Create the plot
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

### Horizontal Bar

```python
# Create the plot
plt.barh(x_axis, y_axis, color='magenta', label='Boston')

# gca means 'get current axes'. We can chain the gca() method to invert_yaxis() to invert the y-axis
plt.gca().invert_yaxis()

# Add legend
plt.legend()

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
```

## Error Bars

```python
# Adding Error Bars
import statistics

stdev = statistics.stdev(y_axis)

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

## Object-Oriented Interface

### Vertical Bar

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')

# Add title
ax.set_title("PyBer Fare by Month")

# Add axes
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

# Add legend
ax.legend()
```

### Horizontal

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

# Invert yaxis
ax.invert_yaxis()

# Add title
ax.set_title("PyBer Fare by Month")

# Add axes
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

# Add legend
ax.legend()
```

```python
# Adding Error Bars
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Major Ticks

```python
# Major ticks
# MATLAB Method
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()

# OOI Method
# ax.xaxis.set_ticks()
# ax.set_xticks()
```

## Minor Ticks

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
