# Create a Pie Chart

## MATLAB Method

```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# To explode a pie wedge, use array-like data points for each value
# autopct- add percentage of each wedge on the pie chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase size of the pie chart
plt.subplots(figsize=(8, 8))

# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')


plt.show()
```

## Object-Oriented Interface

```python
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
color = ['lightcoral', 'peachpuff', 'lemonchiffon', 'palegreen',
         'turquoise', 'lightcyan', 'skyblue', 'lightsteelblue',
         'lavender', 'mediumpurple', 'thistle', 'pink']

fig, ax = plt.subplots(figsize=(8,8))

ax.pie(y_axis,
       explode=explode_values,
       labels=x_axis,
       colors=color,
       autopct='%.1f%%',
       shadow=True,
       startangle=95,
       counterclock=False
      )

plt.show()
```
