# Create a Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## MATLAB Method

```python
# Using plt.plot() method
plt.plot(x_axis, y_axis, 'o')
```

```python
# Using plt.scatter()
plt.scatter(y_axis, x_axis, color='red', label='Chicago')

plt.title("PyBer Fare by Month")

plt.ylabel("Date")
plt.xlabel("Fare($)")

plt.xlim(0, 50)

plt.gca().invert_yaxis()

plt.legend()
```

## Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

# Create a Bubble Chart

## MATLAB Method

```python
# Option 1: Using a for loop
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)

# Can set s=y_axis but bubbles might not be big enough for presentation
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Option 2: Using list comprehension
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Object-Oriented Interface

```python
# alpha- transparency
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color='skyblue', label='Boston', s= [dot_size * 5 for dot_size in y_axis],
          alpha=0.5, edgecolors='black', linewidth=2)

ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.set_xlim(0, 50)

ax.invert_yaxis()

ax.legend()
```
